```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\10725950\Downloads\alerts-data (1).csv")
```

```python
df=df.drop('description',axis='columns')
df=df.drop('status',axis='columns')
df=df.drop('alertId',axis='columns')
df
```

```python
df['lastModifiedTime']=pd.to_datetime(df['lastModifiedTime'])
```

```python
df = df.set_index('lastModifiedTime')
df = df.resample("H").sum()
df
```

```python
data = pd.get_dummies(df)
data
data = data.groupby(['lastModifiedTime']).max()
data
```

```python
for i in range(2,100):
    data = data.replace(i,1)
```

```python
frequent_itemsets = apriori(data,min_support=0.3,use_colnames=True)
rules = association_rules(frequent_itemsets,metric="lift")
rules

```
